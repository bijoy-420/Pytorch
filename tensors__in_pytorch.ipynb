# Reduction operation

```python
e=torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
#sum
torch.sum(e)
#colsum
torch.sum(e,dim=0)
#rowsum
torch.sum(e,dim=1)
```

```python
#mean
torch.mean(e)
#colmean
torch.mean(e,dim=0)
#rowmean
torch.mean(e,dim=1)
```

```python
#median
torch.median(e)
#max and min
torch.max(e)
torch.min(e)
#standrad deviation
torch.std(e)
#product
torch.prod(e)
#variance
torch.var(e)
#argmax(return max value position)
torch.argmax(e)
#argmin(inverse argmax)
torch.argmin(e)
```

# Matrix operations

```python
f=torch.randint(size=(2,3),low=0,high=10)
g=torch.randint(size=(3,2),low=0,high=10)
print(f)
print(g)
```

```python
#matrix multiplication
torch.matmul(f,g)
```

```python
vector1=torch.tensor([1,2,3,])
vector2=torch.tensor([4,5,6])
#dot product
torch.dot(vector1,vector2)
```

```python
#transpose
print(f)
torch.transpose(f,0,1)
```

```python
#determinant
r=torch.randint(size=(3,3),low=0,high=10,dtype=torch.float32)
torch.det(r)
```

```python
#inverse
torch.inverse(r)
```

```python
i=torch.randint(size=(2,3),low=0,high=10)
j=torch.randint(size=(2,3),low=0,high=10)
print(i)
print(j)
```

```python
#greater than
i>j
#less than
i<j
#equal to
i==j
#not equal to
i!=j
```

# Special functions

```python
k=torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
k
```

```python
#log
torch.log(k)
#exp
torch.exp(k)
#sqrt
torch.sqrt(k)
#sin
torch.sin(k)
#cos
torch.cos(k)
#sigmoid
torch.sigmoid(k)
#tanh
torch.tanh(k)
#softmax
torch.softmax(k,dim=1)
#relu(negative value convert to zero)
torch.relu(k)
```
