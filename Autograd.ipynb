# Autograd use for calculate derivation

```python
#In traditional way y=x^2
#derivative dy/dx=2x;
def derivative(x: int):
  return 2*x;
derivative(3)
```

using autograd

```python
import torch
x=torch.tensor(3.0,requires_grad=True)
x
```

```python
y=x**2
y
```

```python
y.backward()
```

```python
x.grad
```

Ex-3

```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
print(w)
print(b)
```

```python
z=w*x+b
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
loss=F.binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
